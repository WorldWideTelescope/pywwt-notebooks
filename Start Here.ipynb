# Example notebooks for pywwt

Welcome! Here you’ll find notebooks for the [pywwt] package, which provides access to the [AAS] [WorldWide Telescope] inside Python and Jupyter environments.

[pywwt]: https://pywwt.readthedocs.org/ 
[AAS]: https://aas.org/
[WorldWide Telescope]: http://www.worldwidetelescope.org/

### Feature Tutorials

These notebooks tour some of the features of [pywwt]:

* [**First Steps**][first-steps] — getting acquainted with the basics of the pywwt viewer
* [**Visualizing Data Tables: the NASA Exoplanet Archive**][nasa-exo] — plotting data in 2D and 3D
* [**Visualizing Imagery: W5 and SN2011fe**][imagery] — visualizing FITS data
* [**The Time Axis: Gamma-Ray Bursts**][grbs] — adding a time axis to your data tables
* [**Adding Annotations**][annots] — including geometrical figures in your WWT displays
* [**Creating Interactive Figures**][int-figs] — export your visualizations to be viewable anywhere!
* [**Visualizing Planetary Data: Earthquakes**][earthquakes] — plotting data on a planetary surface
* [**Animating the Motions of the Solar System Planets in 3D**][sol-sys-sim] — a quick demonstration of the WWT clock
